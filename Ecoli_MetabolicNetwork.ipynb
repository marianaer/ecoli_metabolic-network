Entonces, podemos checar los niveles de fumarato y ATP en condiciones normales.
Naturalmente, la reacción de producción de este metabolito con mayor flujo de este metabolito es SUCDi. Y se utiliza sobre todo en la producción de malato.

```python
# SUCDi es responsable de niveles de fumarato
modelSDH.metabolites.fum_c.summary()
modelSDH.metabolites.atp_c.summary()
modelSDH.summary()
```

También es posible analizar los niveles de metabolitos involucrados con la reacción GND.

GND se encarga de generar D-ribulosa-5-fosfato, necesario para la síntesis de NADPH.

En condiciones normales, la producción de ru5p depende totalmente de GND, con un flujo de 5.2. Se utiliza sobre todo en las reacciones de Ribulosa 5-fosphato 3-epimerasa y Ribosa-5-fosphato isomerasa.

La producción de NADPH depende sobre todo de GND y Glucosa 6-phosphate dehydrogenasa y en mayor medida por Isocitrato Deshidrogenasa (ICDHYr).
Se consume mayormente en la reacción GLUDy (glutamato deshidrogenasa).

La mayoría de la energía en este medio también se destina a la producción de biomasa


```python
# GND es responsable de NADPH, ru5p

modelGND.metabolites.ru5p__D_c.summary()
modelGND.metabolites.nadph_c.summary()
modelGND.metabolites.atp_c.summary()
modelGND.summary()
```

Para cada modelo, se inhiben los genes responsables de las reacciones SUCDi y GND:

```python
# Inhibir los genes a cada modelo
for i in sdh:    
    i.genes.b0721.knock_out()
    i.genes.b0722.knock_out()
    i.genes.b0723.knock_out()
    i.genes.b0724.knock_out()
    
for i in gdn:
    i.genes.b2029.knock_out()
```

Al analizar los niveles de los metabolitos relacionados con SUCDi, observamos que su producción recae mayormente en la reacción ADSL2r, que produce fumarato a partir de 25aics.
La producción de ATP con la inhibición de esta reacción no parece variar mucho respecto a condiciones normales.

La tasa de crecimiento no varía considerablemente entre ambas condiciones (1.1 y 1.09).

```python
modelSDH.metabolites.fum_c.summary()
modelSDH.metabolites.atp_c.summary()
modelSDH.summary()
```

Al analizar los niveles de los metabolitos relacionados con la reacción GND, observamos que en el caso de ru5p, su producción ya no recae en GND, sino en RPE, que lo produce a partir de Xilulosa-5-fosfato, con un flujo de 0.711, mayor al utilizado con la reacción GND.

La producción de NADPH ahora recae en THD2pp, con un flujo de 8.5. Se sabe que esta reacción también es importante en su producción. El flujo de ICDHyr también aumentó de 1.3 a 3.6. Esto probablemente se deba a a un intento de compensarpor la pérdida de la generación de NADPH en oxPPP

El flujo de ATP se mantuvo casi igual (aprox. 69) y se siguió consumiendo en su mayoría en la producción de biomasa.

En este caso, la tasa de crecimiento tampoco varió mucho a las condiciones normales (0.9).

```python
modelGND.metabolites.ru5p__D_c.summary()
modelGND.metabolites.nadph_c.summary()
modelGND.metabolites.atp_c.summary()
modelGND.summary()
```

# Evolucionando las réplicas

Al evolucinoar las cepas con el knockout de la reacción SUCDi, sus tasas de crecimiento se mantuvieron muy similares entre ellas, a la condición normal y a la condición con la reacción knockeada sin evolucionar.
Se mantuvo en valores de ~0.9.

```python
# Optimizando los modelos.
# A los optimizados o evolucionados se les hicieron 3 réplicas

solsSDH=[]
for i in sdh:
    sol=i.optimize()
    solsSDH.append(sol)
    
solsSDH[1:]  # Soluciones de las evolucionadas. (Omitimos la no evolucionada) 
```

Lo mismo sucede en el caso de las cepas inhibidas para la reacción GND:

```python
# Optimizando los modelos KO de GND

solsGND=[]
for i in gdn:
    sol=i.optimize()
    solsGND.append(sol)
    
solsGND[1:]

```

Podemos observar gráficamente los flujos de todo el metabolismo central para las cepas evolucionadas con la deleción de SUCDi:
En las tres cepas, los flujos parecen no variar significativamente entre ellos.
Con respecto a condiciones normales, se vio un aumento en la vía oxPPP y disminución en los flujos de las reacciones correspondientes al ciclo de Krebs, en especial en la reacción AKDGH, que convierte akg a succoa, generando NADH.

Entonces, en este caso, se observa que la tasa de crecimiento no se vio gravemente afectada al hacer los knockouts, pero los flujos de las reacciones y niveles de metabolitos sí cambiaron significativamente
Se observa también un cuello de botella en el ciclo de Krebs, con un alto flujo saliendo del ciclo, relacioinado con la producción de succinato, lo que llevaría a un mayor flujo hacia la asimilación de nitrógeno.

Considerando los resultados obtenidos por el modelo y experimentalmente en el artículo, podría decirse que el modelado metabólico es una aproximación bastante útil y precisa, pues es congruente con lo observado en experimentos. Esto teniendo en cuenta que los modelos sólo consideran alteraciones al metabolismo y no otros tipos de regulación que sí suceden en la célula, lo cual podría explicar que no todas las reacciones y sus cambios están representadas en ambas redes. Es también interesante cómo suele ser más o menos congruente un modelo "optimizado" con el estado de la célula después de la evolución; es decir, que la célula llegue al estado más óptimo posible teniendo en cuenta las diferentes condiciones ambientales y de su metabolismo.
Un método de COBRA que podría utilizarse con lo obtenido en el experimento podría ser OptKnock, pues sería interesante identificar procesos reguladores que conntribuyan al desarrollo del crecimiento óptimo en cepas evolucionadas.
